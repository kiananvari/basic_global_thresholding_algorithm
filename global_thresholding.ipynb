```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
image = cv.imread("./test.png")
new_image = image.copy()
plt.imshow(image, cmap="gray")
```

```python
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(histogram, color='k')
plt.show()
```

```python
threshold = image.mean()
print(f"First threshold(mean) = {threshold}")


flag = True
while(flag):
    group1 = image[image<threshold].mean()
    group2 = image[image>threshold].mean()
    new_threshold = (group1 + group2)/2
    print(f"new Threshold = {new_threshold}")

    if abs(new_threshold-threshold) < 5:
        flag = False
        
    threshold = new_threshold

```

```python
print(f"Threshold = {threshold}")

new_image[new_image<threshold] = 0
new_image[new_image>threshold] = 255
```

```python
histogram = cv2.calcHist([new_image], [0], None, [256], [0, 256])
plt.plot(histogram, color='k')
plt.show()
```
